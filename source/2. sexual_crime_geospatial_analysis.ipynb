```python
import geopandas as gpd

bogota_gdf = gpd.read_file("../data/poligonos-localidades.geojson")
bogota_gdf.head(20)
```

```python
bogota_crime_gdf = gpd.read_file("../data/DAILoc.geojson")
bogota_crime_gdf.head(20)
```

```python
import ast
def extract_latitude(geo_point):
    geo_dict = ast.literal_eval(geo_point)  
    return geo_dict['lat']

def extract_longitude(geo_point):
    geo_dict = ast.literal_eval(geo_point)  
    return geo_dict['lon']

# Apply the function to create a new 'latitude' column

bogota_gdf['latitude'] = bogota_gdf['geo_point_2d'].apply(extract_latitude)
bogota_gdf['longitude'] = bogota_gdf['geo_point_2d'].apply(extract_longitude)
```

```python
import folium
from folium.plugins import HeatMapWithTime
from IPython.display import IFrame
import pandas as pd
import numpy as np

# Function to embed the map
def embed_map(m, file_name):
    m.save(file_name)
    return IFrame(file_name, width='100%', height='500px')

# Merge the GeoDataFrames on the "CMIULOCAL" column
bogota_gdf = bogota_gdf.rename(columns={"Identificador unico de la localidad": "CMIULOCAL"})
bogota_crime_gdf = bogota_crime_gdf.merge(bogota_gdf, on="CMIULOCAL")

# List of years to normalize
years = ['CMDS18CONT', 'CMDS19CONT', 'CMDS20CONT', 'CMDS21CONT', 'CMDS22CONT', 'CMDS23CONT']

# Find the global maximum across all years
global_max = bogota_crime_gdf[years].max()

# Normalize each year's data by the global maximum
for year in years:
    global_max = bogota_crime_gdf[year].max()
    bogota_crime_gdf[year +"_1"] = bogota_crime_gdf[year] / global_max

# Prepare the data for HeatMapWithTime
heatmap_data = []
for year in years:
    data = bogota_crime_gdf[['latitude', 'longitude', year +"_1"]].dropna().values.tolist()
    heatmap_data.append(data)

# Create a map centered around Bogotá (approximate coordinates)
m = folium.Map(location=[4.60971, -74.08175], zoom_start=11)

# Add time-enabled heatmap
#tell me more parameters of heatmapwithtime  to gaussian blur, radius, etc
#
#https://python-visualization.github.io/folium/plugins.html
 
 
HeatMapWithTime(heatmap_data, index=[2018, 2019, 2020, 2021, 2022, 2023], auto_play=True, max_opacity=0.8).add_to(m)

for index, row in bogota_crime_gdf.iterrows():
    for i, year in enumerate(years):
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=0.5,
            color='blue',
            fill=False,
            fill_color='blue',
            fill_opacity=0.1
            #popup=folium.Popup(f"{row['CMNOMLOCAL']}<br>{'Cases'}: {row[year]:.2f}", max_width=300)
        ).add_to(m)

for index, row in bogota_crime_gdf.iterrows():
    for i, year in enumerate(years):
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            icon=folium.DivIcon(html=f"""<div style="font-size: 6px; color: black;">
                                         {row['CMNOMLOCAL']}</div>""")
        ).add_to(m)
# Embed the map
embed_map(m, 'bogota_sex_crimes_heatmap_time_global_normalized.html')
```
